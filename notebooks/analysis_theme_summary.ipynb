```python
# analysis_theme_summary.ipynb

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your theme summary CSV
df = pd.read_csv("../data/theme_summary.csv")

#Check the data
df.head()
```

```python

# Create a bar plot
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x="bank",
    y="count",
    hue="theme",
    palette="Set2"
)

plt.title("Theme Distribution per Bank")
plt.xlabel("Bank App")
plt.ylabel("Number of Keywords")
plt.xticks(rotation=30)
plt.legend(title="Theme", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("../visualizations/theme_distribution_bar.png", dpi=300, bbox_inches='tight')

plt.show()
```
